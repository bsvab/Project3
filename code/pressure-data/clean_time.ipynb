```python
# Import dependencies
import pandas as pd
import os
from datetime import datetime, timedelta
import pyproj
import numpy as np
```

```python
# Get the directory of the current script
current_directory = os.getcwd()
# Name of your CSV file
filename = 'cleaned_pressure_data.csv'

# Construct the file path using os.path.join
file_path = os.path.join(current_directory, filename)
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)
```

```python
# Display the first and last 5 rows of the new DataFrame
df.head(-5)
```

```python
# Convert time to dates
df['Time'] = pd.to_datetime('1902-12-01') + pd.to_timedelta(df['Time'], unit='D')
```

```python
# Display the DataFrame with updated Time Column
df
```

```python
# Export the DataFrame to a csv file
df.to_csv('updated_pressure_data.csv', index=False)
```
