```python
# Import dependencies
import pandas as pd
import os
import pyproj
import numpy as np
from datetime import datetime, timedelta
```

```python
# Get the directory of the current script
current_directory = os.getcwd()
# Name of your CSV file
filename = 'output.csv'

# Construct the file path using os.path.join
file_path = os.path.join(current_directory, filename)
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)
```

```python
# Display the first 5 rows of the new DataFrame
df.head()
```

```python
df = df.rename(columns={
    'Time' : 'Time',
    'X' : 'X',
    'Y' : 'Y',
    'Layer' : 'Pressure',
    'Layer Number' : 'Layer'
})
```

```python
df.head()
```

```python
# Duplicate the dataframe
df_dupe = df
```

```python
df_dupe.head()
```

```python
# Initialize number variable
number = None

# Iterate over each row
for index, row in df_dupe.iterrows():
    if row['Pressure'].startswith('Layer_'):
        number = row['Pressure']

    df_dupe.at[index, 'Layer'] = number
```

```python
df_dupe.head()
```

```python
# Duplicate the DataFrame
df_layers = df_dupe
```

```python
# Filter the DataFrame to remove rows where 'X' column is equal to 'X'
df_filtered = df_layers[df_layers['X'] != 'X']
```

```python
df_filtered.head()
```

```python
# Export final CSV file
df_filtered.to_csv('cleaned_pressure_data.csv', index=False)
```
